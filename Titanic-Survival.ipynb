# Data Cleaning & Preprocessing

```python
# Fill missing age with the median
data['Age'].fillna(data['Age'].median(), inplace=True)
data
```

```python
# Fill missing fare with the median
data['Fare'].fillna(data['Fare'].median(), inplace=True)
data
```

```python
# Fill missing embarked with the most frequent value
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
data
```

```python
# Drop unnecessary columns
data.drop(['Cabin', 'Ticket', 'PassengerId', 'Name'], axis=1, inplace=True)
data
```

```python
# Encode categorical variables
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
data
```

```python
data['Embarked'] = data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
data
```

```python
# Normalize a numerical data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data[['Age', 'Fare']] = scaler.fit_transform(data[['Age', 'Fare']])
data.head()
```

# Split the Data

```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop('Survived', axis=1)
y = data['Survived']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x
```

```python
y
```

# Model Training

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

# Evaluate the Model

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
print(" Accuracy:", accuracy_score(y_test, y_pred))
print("\n Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Plot confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, cmap='Purples', fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

#  Save the Model

```python
import joblib

# Save the model to the models folder
joblib.dump(model, 'models/titanic_model.pkl')
print("\n Model saved as 'titanic_model.pkl'")
```
